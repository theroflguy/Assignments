```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
mymovies = pd.read_csv("/media/gargi/Data/DataSets/my_movies.csv")
mymovies.head()
```

# Data Preprocessing 
as the data is not in transaction form

```python
df=pd.get_dummies(mymovies)
df.head()
```

# APRIORI ALGORITHM

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
#####  An leverage value of 0 indicates independence. Range will be [-1 1]
# high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
